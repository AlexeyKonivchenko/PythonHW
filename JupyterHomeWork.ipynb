```python
from matplotlib import pyplot as plt

x = [1, 2]
y = [3, 4]
plt.plot(x,y)
```

```python
# x - диапазон -100 100
# y - 
x = [x for x in range(-100,100)]
y = [t**3 for t in x]
plt.plot(x,y)
```

```python
# x - диапазон -100 100
# y - 
x = [x+0.5 for x in range(-100,100)]
y = [1/t for t in x]
plt.plot(x,y)
```

```python
import math
def f(x):
    return math.sin(x) + 5*x
x = [x/100 for x in range(-300,300)]
y = [f(t) for t in x]
plt.plot(x,y)
```

```python
def f(x):
    return 3*x*x -4*x +10
x = [x for x in range(-30,30)]
y = [f(t) for t in x]
plt.plot(x,y)
```
